## 9) Make a histogram the distribution of the Earthquake magnitudes

#### Do one subplot for the filtered and one for the unfiltered dataset. Use a Logarithmic scale. What sort of relationship do you see?

```python
df_filtered = df[df['mag']>=4]
```

```python
fig,axe = plt.subplots(1,2,figsize=(12,5))
axe[0].hist(df['mag'],bins=50)
axe[0].set_yscale('log')
axe[0].set_title('All Earthquakes')
axe[0].set_xlabel('Magnitude')
axe[0].set_ylabel('Count (log scale)')

axe[1].hist(df_filtered['mag'],bins=50, range=(4,10), rwidth=0.8)
axe[1].set_yscale('log')
axe[1].set_title('Earthquakes (mag ≥ 4)')
axe[1].set_xlabel('Magnitude')
axe[1].set_ylabel('Count (log scale)')
axe[1].set_xlim((4, 10))
```

### What sort of relationship do you see?
Both histograms show a strongly skewed distribution, with many more small earthquakes than large ones. The filtered dataset (mag ≥ 4) represents a subset of the full dataset and follows a similar overall pattern.
In the full dataset, the peak occurs around magnitude 1, indicating that small-magnitude earthquakes are the most frequent. In the filtered dataset, the peak shifts to around magnitude 4–5, since lower-magnitude events have been removed. The frequency of earthquakes decreases rapidly as magnitude increases, showing that large earthquakes are much rarer than smaller ones.
